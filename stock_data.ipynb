# Project: Stock Data

XTB and PKN Orlen stock price data collected from yfinance api.

XTB:

```python
import yfinance as yf
xtb = yf.Ticker('XTB.WA')
xtb_data = xtb.history(period='max')
xtb_data.head()
```

PKN Orlen:

```python
import yfinance as yf
pkn = yf.Ticker('PKN.WA')
pkn_data = pkn.history(period='max')
pkn_data.head()
```

XTB and PKN Orlen revenue data collected from official company sites.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'PKN_Orlen': {
        '2005': 41.19, '2006': 52.87, '2007': 63.79, '2008': 79.53,
        '2009': 67.93, '2010': 83.55, '2011': 106.97, '2012': 120.10,
        '2013': 113.85, '2014': 106.83, '2015': 88.34, '2016': 79.55,
        '2017': 95.36, '2018': 109.71, '2019': 111.20, '2020': 86.18,
        '2021': 131.34, '2022': 282.42, '2023': 372.77, '2024': 296.95
    },
    'XTB': {
        '2016': 0.250, '2017': 0.289, '2018': 0.308,
        '2019': 0.238, '2020': 0.799, '2021': 0.626,
        '2022': 1.450, '2023': 1.618, '2024': 1.873
    }
}

df_pkn = pd.DataFrame.from_dict(data['PKN_Orlen'], orient='index', columns=['Revenue'])
df_xtb = pd.DataFrame.from_dict(data['XTB'], orient='index', columns=['Revenue'])

plt.figure(figsize=(12, 6))

plt.plot(df_pkn.index, df_pkn['Revenue'],
         marker='o', color='red', linewidth=2, label='PKN Orlen (mld PLN)')

plt.title('Revenue PKN Orlen (2005-2023)', fontsize=14, pad=20)
plt.xlabel('Year', labelpad=10)
plt.ylabel('Revenue (mld PLN)', labelpad=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

for year, rev in df_pkn['Revenue'].items():
    plt.text(year, rev, f'{rev:.1f}', ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.plot(df_xtb.index, df_xtb['Revenue'],
         marker='s', color='blue', linewidth=2, label='XTB (mld PLN)')

plt.title('Revenue XTB (2016-2023)', fontsize=14, pad=20)
plt.xlabel('Year', labelpad=10)
plt.ylabel('Revenue (mld PLN)', labelpad=10)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

for year, rev in df_xtb['Revenue'].items():
    plt.text(year, rev, f'{rev:.1f}', ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Charts for XTB showing revenue and stock price.

```python
plt.plot(xtb_data.index, xtb_data['Close'], label='Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Stock Price (PLN)')
plt.title('XTB Stock Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

Charts for PKN Orlen showing revenue and stock price.

```python
plt.plot(pkn_data.index, pkn_data['Close'], label='Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Stock Price (PLN)')
plt.title('PKN Orlen Stock Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

How revenue affects stock prices.

```python
df_pkn.index = pd.to_datetime(df_pkn.index)
df_xtb.index = pd.to_datetime(df_xtb.index)

plt.figure(figsize=(12, 6))

plt.plot(df_xtb.index, df_xtb['Revenue'],
         marker='o', color='red', linewidth=2, label='Revenue (billion PLN)')
plt.ylabel('Revenue (billion PLN)', color='red')
plt.xlabel('Year')
plt.title('XTB: Revenue and Stock Price', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

ax2 = plt.twinx()
ax2.plot(xtb_data.index, xtb_data['Close'],
         color='blue', linewidth=2, label='Stock Price (PLN)')
ax2.set_ylabel('Stock Price (PLN)', color='blue')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

plt.plot(df_pkn.index, df_pkn['Revenue'],
         marker='o', color='red', linewidth=2, label='Revenue (billion PLN)')
plt.ylabel('Revenue (billion PLN)', color='red')
plt.xlabel('Year')
plt.title('PKN Orlen: Revenue and Stock Price', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

ax2 = plt.twinx()
ax2.plot(pkn_data.index, pkn_data['Close'],
         color='blue', linewidth=2, label='Stock Price (PLN)')
ax2.set_ylabel('Stock Price (PLN)', color='blue')

plt.tight_layout()
plt.show()
```
